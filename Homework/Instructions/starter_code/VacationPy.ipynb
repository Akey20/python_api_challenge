### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
cities_df = pd.read_csv('../output_data/city_df.csv')

# Get rid of unnecessary extra index
del cities_df['Unnamed: 0']

# Show all rows of the DataFrame  - this is how I like to see it can comment out 
pd.set_option('display.max_rows', None)

cities_df
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
gmaps.configure(api_key=g_key)
#convert humidity to float / dropna for NaN values
cities_df = cities_df.dropna()
humidity = cities_df["Humidity"].astype(float)

#lat, long for Locations
locations = cities_df[["Lat", "Long"]].astype(float)

#humidity heatmap
fig = gmaps.figure(center = [0,0], zoom_level = 2)

#heat layer # I removed max intensity = 100 because it wasn't a good look
heat_layer = gmaps.heatmap_layer(locations, weights = humidity, dissipating = False,  
                                point_radius =4)
fig.add_layer(heat_layer)
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
weather_df = cities_df

#filter for desired conditions 
weather_df = weather_df.loc[(weather_df["Max Temp"] < 85) & (weather_df["Max Temp"] > 65)]
weather_df = weather_df.loc[(weather_df["Wind"] < 10)] 
weather_df = weather_df.loc[(weather_df["Cloudiness"] < 10)] 
weather_df = weather_df.loc[(weather_df["Humidity"] < 65)]  # this is my preference but drops city count can be removed

#drop null values
weather_df = weather_df.dropna()

#check dataframe
pd.set_option('display.max_rows', None)
weather_df  # .count() - used at end just to see count remove for production and just leave weather_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df = weather_df

#params
params = {"radius": 5000,
         "types": "lodging", 
         "key": g_key}
for index, row in hotel_df.iterrows():
    lat = row["Lat"]
    long = row["Long"]
    
    params["location"] = f"{lat},{long}"
    
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
  
    #url & api request
    print(f"Getting Results {index}: {row['City']}.")
    response = requests.get(base_url, params = params).json()
    
    # extract info
    results = response['results']
    
    try:
        print(f"Nearest hotel is {results[0]['name']}.")
        hotel_df.loc[index, 'Hotel Name'] = results[0]['name']
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        
    print("-------------------------------")
```

```python
# NOTE: Do not change any of the code in this cell
locations = hotel_df[["Lat", "Long"]]
markers = gmaps.marker_layer(locations)
# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
#locations = hotel_df[["Lat", "Lng"]]
hotel_info
```

```python
# Add marker layer ontop of heat map
hotel_layer = gmaps.symbol_layer(
    locations, fill_color='rgba(0, 150, 0, 0.4)',
    stroke_color='rgba(0, 0, 150, 0.4)', scale=2,
    info_box_content=hotel_info)

# Display figure
fig.add_layer(markers)
fig.add_layer(hotel_layer)
fig
```
